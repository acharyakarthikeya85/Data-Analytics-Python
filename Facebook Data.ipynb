## Data Wrangling

### Missing value imputation

```python
facebook_df['gender'].mode()
```

```python
# Since gender is a categorical data we fill missing values with mode instead of mean/median 
facebook_df['gender'].fillna(str(facebook_df['gender'].mode().values[0]), inplace=True)
```

```python
# Difference between max and 75% is more comapred to min and 25% for the tenure column.
# Hence we fill missing values with median instead of mean. 
# Standard deviation is high which indicates that there are outliers in the data. Mean is drawn towards outliers.
```

```python
Group_by_gender = facebook_df.groupby('gender')
# print(Group_by_gender)
print(Group_by_gender.median())
```

```python
def impute_with_median(series):
    return series.fillna(series.median())

facebook_df.tenure = Group_by_gender['tenure'].transform(impute_with_median)
```

## Data transformation

```python
#Transform gender categorical data to numerical values
def numerical_gender(gender):
    if(gender == 'male'):
        gender = 1
    else:
        gender = 0
    return gender    

facebook_df['gender'] = facebook_df['gender'].apply(numerical_gender)
```

```python
facebook_df.head()
```

### Create a new column called Facebook_Usertype

```python
def user_type(gender_with_age):
    gender, age = gender_with_age
    if(age < 20):
        return 'Teenager'
    else:
        if(gender == 1):
            return 'Male'
        else:
            return 'Female'
        
facebook_df['Facebook_Usertype'] = facebook_df[['gender','age']].apply(user_type, axis=1)         
```

```python
facebook_df.tail(30)
```

## Feature engineering

```python
# We can convert the categorical user type of facebook to numerical_type for further analysis.
def Fb_user_type(user_type):
    if(user_type == 'Teenager'):
        uType = 2
    elif(user_type == 'Male'):
        uType = 1
    else: 
        uType = 0
        
    return uType     

facebook_df['FBUserType_num'] = facebook_df['Facebook_Usertype'].apply(Fb_user_type)
```
